```python
track_list = df['Track Name'].unique().tolist()
```

```python
def track_to_region(track_name):
    if track_name in track_list:
        return dict((x,y) for x,y in tuple(zip(df[df['Track Name'] == track_name].groupby(['Region']).sum().index.tolist(),df[df['Track Name'] == track_name].groupby(['Region']).sum().values[:,1].tolist())))
    else:
        print('Track not found')  
```

```python
def track_to_month(track_name):
    if track_name in track_list:
        return dict((x,y) for x,y in tuple(zip(df[(df['Track Name'] == track_name) & (df['Year'] == '2017')].groupby(['Month_name']).sum().index.tolist(),df[(df['Track Name'] == track_name) & (df['Year'] == '2017')].groupby(['Month_name']).sum().values[:,1])))
    else:
        print('Track not found')
```

```python
df['Year'] = df['Date'].apply(lambda x : x.split('-')[0])
df['Date'] = pd.to_datetime(df['Date'])
df['Month_name'] = df['Date'].dt.month_name()
```
